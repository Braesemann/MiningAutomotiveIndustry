## Degree Distribution

## Toyota

```python
fh=open("results/toyota_edges.csv", 'rb')
G=nx.read_edgelist(fh, delimiter=",")
fh.close()

G.remove_edges_from(G.selfloop_edges())
rich_club = list(nx.rich_club_coefficient(G,normalized=False).values())

print(np.mean(rich_club))
```

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

degree_distribution = plt.figure()
#plt.loglog(deg,cnt, c = 'midnightblue', marker='o', markersize=4, linestyle = 'None',)

logx = np.log(deg)
logy = np.log(cnt)
coeffs = np.polyfit(logx,logy,deg=3)
poly = np.poly1d(coeffs)

yfit = lambda deg: np.exp(poly(np.log(deg)))
                          
#plt.loglog(deg,yfit(deg), 'midnightblue' , linewidth=6, markersize=6)

x1 = range(len(cnt))
markers = np.power(x1, 0.5)
#print(markers)

plt.loglog(deg,yfit(deg),  c = 'midnightblue' , linewidth=20, markersize=6)


degree_distribution = plt.figure()
for i in range(0, len(cnt)):
    
    plt.loglog(deg[i],cnt[i], c = 'midnightblue', marker='o', markersize=markers[i], linestyle = 'None',)




## TextBox
textstr = '\n'.join((
    "Average Degree: 3.59",
    "Average Rich Club Coefficient: 0.23",
    "Network Density: 0.01",
    "Network Diameter: 8"))

props = dict(boxstyle='round', facecolor='gray', alpha=0.1)

# place a text box in upper left in axes coords
#plt.text(11.00, 6500.00, textstr, fontsize=10, fontweight='bold',
#        verticalalignment='top', bbox=props)

plt.grid(which='major', axis='y', linestyle='--')
plt.ylabel("P(k)", fontsize=12, fontweight='bold')
plt.xlabel("Node Degree k", fontsize=12, fontweight='bold')
plt.show()
```

## Webpage Analytics

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties



fig3 = plt.figure()


data = [[ 0.086, 0.139, 0.008],
        [ 0.088, 0.041,  0.085],
        [ 0.554, 1.057,  0.179],
        [ 0.089, 0.104, 0.127],
        [ 0.825, 0.066,  0.262],
        [ 0.835, 0.153,  0.567]]


columns = ('Toyota', 'VW', 'Hyundai')
rows = ["% e-mobility", "% connectivity", "% autonom. driving","aver. page response time", "aver. number of videos", "aver. social media connectivity"]


values = np.arange(0, 2.5, 0.25)
value_increment = 1.0

# Get some pastel shades for the colors
colors = plt.cm.PuBu(np.linspace(0.4, 0.8, len(rows)))
n_rows = len(data)

index = np.arange(len(columns)) + 0.3
bar_width = 0.4

# Initialize the vertical-offset for the stacked bar chart.
y_offset = np.zeros(len(columns))

# Plot bars and create text labels for the table
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    new_offset = np.zeros(len(columns))
    y_offset = y_offset + new_offset + data[row]
    cell_text.append([(round(x,3)) for x in data[row]])
    
    
# Reverse colors and text labels to display the last value at the top.
colors = colors[::-1]
cell_text.reverse()

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns,
                      loc='bottom')

# make text bold
for (row, col), cell in the_table.get_celld().items():
    if (row == 0) or (col == -1):
        cell.set_text_props(fontproperties=FontProperties(weight='bold'))
        

# Adjust layout to make room for the table:
plt.subplots_adjust(left=0.4, bottom=0.25)

plt.ylabel("aggregated ratio",fontweight='bold')
plt.yticks(values * value_increment, [val for val in values])
plt.xticks([])
plt.title('Digital Trend Analysis', fontweight='bold')

plt.show()
```

## Node Degree and Trends

```python
domain_name = "toyota"
```

```python
import matplotlib.pylab as plt
import numpy as np
import collections
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.collections import LineCollection
import csv



fh=open('results/' + domain_name + '_edges.csv', "rb")
G=nx.read_edgelist(fh, delimiter=",")
fh.close()

## remove edges
G.remove_edges_from(G.selfloop_edges())
```

## K-Core

```python

## get k-core
kcore = nx.k_core(G, k=None)
print("k-core length: ", len(kcore))
print("k-core nodes: ", kcore.nodes)
```

## PageRank

```python
page_ranks = nx.pagerank(G, alpha=0.85, personalization=None, max_iter=100, tol=1e-06, nstart=None, weight='weight', dangling=None)

print(page_ranks)
```

## Node Degree

```python
# get node degree
node_degrees = G.degree

print(node_degrees)
```

## Write to Excel File

```python
## Reader
nodes_csv = open('results/' + domain_name + '_nodes_tags.csv')
reader = csv.DictReader(nodes_csv)

## Writer
nodes_tags_csv = open('results/' + domain_name + '_nodes_tags_2.csv', "w")
#headers
header_list = list(next(reader).keys()) + ["node_degree"] + ["page_rank"]
header_string = ','.join(map(str, header_list)) 
nodes_tags_csv.write(str(header_string) + '\n')


for row in reader:
    
    webpage = row['Address']
    
    # node degree
    degree = node_degrees[webpage]

    
    # page rank 
    page_rank = round(page_ranks[webpage],8)
    
        
    row_string = ','.join(map(str, row.values()))
        
    
    ## id, label, address, titel, ....., videos, social media, emobility .... sentiment, language
    nodes_tags_csv.write(row_string + "," + str(degree) + "," + str(page_rank) + "\n")

print("\n\n analysis terminated")
    
nodes_csv.close()
nodes_tags_csv.close()
```
